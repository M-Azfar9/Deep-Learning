```python
import pandas as pd
```

```python
data = pd.read_csv('DailyDelhiClimateTest.csv')[['date', 'meantemp']]
```

```python
import matplotlib.pyplot as plt

plt.scatter(data['date'], data['meantemp'])
plt.show()
```

> ### Calculating EWMA
- alpha is opposite to beta
- if we set alpha to 0.9, then beta is 0.1

```python
data['ewma']=data['meantemp'].ewm(alpha=0.1).mean()
```

```python
plt.scatter(data['date'], data['meantemp'], color='red')
plt.plot(data['date'], data['meantemp'], color='blue',label='original')
plt.plot(data['date'], data['ewma'], color='black',label='ewma')
plt.legend()
plt.show()
```
