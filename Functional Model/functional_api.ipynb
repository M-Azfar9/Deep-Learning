  <ul>

  </ul>

  <p>The <strong>Keras Functional Model</strong> refers to models built using the Functional API. It allows you to define models as directed acyclic graphs (DAGs) of layers, instead of just a linear stack.</p>

  <p>ðŸ‘‰ With it, you can build:</p>

  <ul>
    <li>Multi-input models</li>
    <li>Multi-output models</li>
    <li>Models with shared layers</li>
    <li>Models with non-linear layer connections (e.g., skip connections like in ResNet)</li>
  </ul>

```python
from tensorflow.keras.models import Model
from tensorflow.keras.layers import * 
```

```python
# here you tell, which layer is input layer and which is output layer
model = Model(inputs=x, outputs=[output1, output2],)
```

Building our architecture

```python
# 3 columns input
x = Input(shape=(3,))
```

```python
# its mean this hidden layer got input from x, or connected with x
hidden1 = Dense(128, activation='relu')(x)
# similarly
hidden2 = Dense(64, activation='relu')(hidden1)

# to predict age
output1 = Dense(1, activation='linear')(hidden2)
# to predict city
output2 = Dense(1, activation='sigmoid')(hidden2)
```

```python
model.summary()
```

```python
from keras.utils import plot_model
plot_model(model)
```

Another model

```python
inputA  = Input(shape=(32, ))
inputB  = Input(shape=(128, ))
```

```python
x = Dense(8, activation='relu')(inputA)
x1 = Dense(4, activation='relu')(x)
```

```python
y = Dense(64, activation='relu')(inputB)
y1 = Dense(32, activation='relu')(y)
y2 = Dense(4, activation='relu')(y1)
```

```python
# combined the output of the two layers
combined = concatenate([x1, y2])
```

```python
# apply a fc layer
z = Dense(2, activation='relu')(combined)
z1 = Dense(1, activation='linear')(z)
```
