> # ***HE INITIALIZATION*** 

```python
from sklearn.datasets import make_moons
x, y = make_moons(n_samples=200, noise=0.2)
```

```python
import matplotlib.pyplot as plt
plt.scatter(x[:,0], x[:,1], c=y)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
model = Sequential([
    # Dense(30, input_dim=2, activation='sigmoid', kernel_initializer='glorat_normal'),
    # Dense(30, input_dim=2, activation='relu', kernel_initializer='glorat_uniform'),
    Dense(30, input_dim=2, activation='relu', kernel_initializer='he_normal'),
    Dense(30, input_dim=2, activation='relu', kernel_initializer='he_normal'),
    Dense(30, input_dim=2, activation='relu', kernel_initializer='he_normal'),
    Dense(1, activation='sigmoid')
])
model.summary()
```

```python
model.get_weights()
```

```python
model.get_weights()
```

```python
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
history = model.fit(x, y, epochs=200, validation_split=0.2, verbose=1)
```

```python
# for tanh and relu it remain same as 0
# but in sigmoid the all weights release from a node has same value, so it works like a single neurons
model.get_weights()
```

```python
from mlxtend.plotting import plot_decision_regions
plot_decision_regions(x, y, clf=model, legend=2)
```
