```python
from sklearn.datasets import make_moons
x, y = make_moons(n_samples=200, noise=0.2)
```

```python
import matplotlib.pyplot as plt
plt.scatter(x[:,0], x[:,1], c=y)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
model = Sequential([
    # Dense(30, input_dim=2, activation='sigmoid'),
    # Dense(30, input_dim=2, activation='relu'),
    Dense(30, input_dim=2, activation='tanh'),
    Dense(1, activation='sigmoid')
])
model.summary()
```

```python
model.get_weights()
```

```python
initial_weigths = model.get_weights()
```

```python
import numpy as np
```

```python
# for weights 0
initial_weigths[0] = np.zeros(initial_weigths[0].shape)
initial_weigths[1] = np.zeros(initial_weigths[1].shape)
initial_weigths[2] = np.zeros(initial_weigths[2].shape)
initial_weigths[3] = np.zeros(initial_weigths[3].shape)
```

```python
# for random weights
for i in range(0,4):
    initial_weigths[i] = np.random.randn(*initial_weigths[i].shape) * 0.01
```

```python
initial_weigths
```

```python
model.set_weights(initial_weigths)
```

```python
model.get_weights()
```

```python
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
history = model.fit(x, y, epochs=200, validation_split=0.2, verbose=1)
```

```python
# for tanh and relu it remain same as 0
# but in sigmoid the all weights release from a node has same value, so it works like a single neurons
model.get_weights()
```

```python
from mlxtend.plotting import plot_decision_regions
plot_decision_regions(x, y, clf=model, legend=2)
```
