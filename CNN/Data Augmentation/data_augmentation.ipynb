```python
from keras.preprocessing import image
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```python
img = image.load_img(r'D:\Gen Ai\Deep Learning\CNN\Data Augmentation\cat2.jpg')
```

```python
img
```

```python
datagen = ImageDataGenerator(
    rotation_range = 30,
    shear_range = 0.2, 
    zoom_range = 0.2, 
    horizontal_flip = True, 
    vertical_flip = True,
    width_shift_range = 0.2,
    height_shift_range = 0.2, 
)
```

```python
img_vector = image.img_to_array(img)
```

```python
img_vector.shape
```

```python
input_batch = img_vector.reshape(1, 148, 184, 3)
```

```python
i = 0

for output in datagen.flow(input_batch, batch_size=1, save_to_dir='augmented_data'):
    i = i+1
    if i == 10:
        break
```
