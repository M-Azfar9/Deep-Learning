> # Perceptron implementation using GD and Loss Function

```python
from sklearn.datasets import make_classification
x, y = make_classification(n_samples=200, n_classes=2, n_clusters_per_class=1, n_features=2, n_informative=1, n_redundant=0, random_state=41, hypercube=False, class_sep=15)
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
plt.scatter(x[:,0], x[:,1], c=y)
```

```python
def perceptron(x, y):
    w1 = w2 = b = 1
    lr = 0.1
    epochs = 1000
    for j in range(epochs):
        # using stochastic GD
        for i in range(x.shape[0]):
            z = w1*x[i][0] + w2*x[i][1] + b
            
            if z*y[i] < 0:
                w1 = w1 + lr*y[i]*x[i][0]
                w2 = w2 + lr*y[i]*x[i][1]
                b = b + lr*y[i]
    
    return w1, w2, b
```

```python
w1, w2, b = perceptron(x, y)
```

```python
m = - (w1/w2)
c = - (b/w2)
print(m, c)
```

```python
x_input = np.linspace(-3, 3, 100)
```

```python
y_input = m*x_input + c
```

```python
plt.plot(x_input, y_input, color='red')
plt.scatter(x[:,0], x[:,1], c=y)
plt.ylim(-3, 2)
```
