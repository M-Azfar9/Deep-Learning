> # ***BATCH NORMALIZATION*** 

```python
from sklearn.datasets import make_moons
x, y = make_moons(n_samples=200, noise=0.2)
```

```python
import matplotlib.pyplot as plt
plt.scatter(x[:,0], x[:,1], c=y)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, BatchNormalization
```

```python
model = Sequential([
    Dense(30, input_dim=2, activation='relu'),
    Dense(30, input_dim=2, activation='relu'),
    Dense(30, input_dim=2, activation='relu'),
    Dense(1, activation='sigmoid')
])
model.summary()
```

```python
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
history = model.fit(x, y, epochs=200, validation_split=0.2, verbose=1)
```

```python
# for tanh and relu it remain same as 0
# but in sigmoid the all weights release from a node has same value, so it works like a single neurons
model.get_weights()
```

```python
from mlxtend.plotting import plot_decision_regions
plot_decision_regions(x, y, clf=model, legend=2)
```

```python
model_with_batch_norm = Sequential([
    Dense(30, input_dim=2, activation='relu'),
    BatchNormalization(),
    Dense(30, input_dim=2, activation='relu'),
    BatchNormalization(),
    Dense(30, input_dim=2, activation='relu'),
    BatchNormalization(),
    Dense(1, activation='sigmoid')
])
model_with_batch_norm.summary()
```

```python
model_with_batch_norm.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
history_of_batch_norm = model_with_batch_norm.fit(x, y, epochs=200, validation_split=0.2, verbose=1)
```

```python
from mlxtend.plotting import plot_decision_regions
plot_decision_regions(x, y, clf=model_with_batch_norm, legend=2)
```

```python
plt.plot(history.history['accuracy'])
plt.plot(history_of_batch_norm.history['accuracy'], color='green')
```

```python
plt.plot(history.history['val_accuracy'])
plt.plot(history_of_batch_norm.history['val_accuracy'], color='green')
```
